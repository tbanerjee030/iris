```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris = iris[['petal_length','petal_width']]
```

```python
iris
```

```python
x = iris['petal_length']
y = iris['petal_width']
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("petal length")
plt.ylabel("petal width")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.4, random_state = 23)
```

```python
import numpy as np
x_train = np.array(x_train).reshape(-1,1)
x_train
```

```python
x_test = np.array(x_test).reshape(-1,1)
x_test.flatten()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
y_pred_train = m*x_train + c # instead of this formula we can use the predict function which is done in the next block
y_pred_train.flatten()
```

```python
y_pred_train1 = lr.predict(x_train)
y_pred_train1
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train)
plt.plot(x_train,y_pred_train1, color = 'red')
plt.xlabel("petal length")
plt.ylabel("petal width")
```

```python
y_pred_test1 = lr.predict(x_test)
y_pred_test1
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred_test1, color = 'orange')
plt.xlabel("petal length")
plt.ylabel("petal width")
```
